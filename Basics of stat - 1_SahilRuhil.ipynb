# 2) Data visualization

```python
# Bar Chart Analysis for Categorical Column
```

```python
# Selecting categorical columns
df_cat = df.select_dtypes(exclude=['int','float'])
```

```python
df_cat
```

```python
# Plotting the Brand distribution

df_cat['Brand'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Brands')
plt.xlabel('Brand')
plt.ylabel('Count of Transactions')
plt.show()
```

```python
# Plotting the model distribution

df_cat['Model'].value_counts().plot(kind='bar', color='coral', edgecolor='black')
plt.title('Distribution of Models')
plt.xlabel('Model')
plt.ylabel('Count of Transactions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the city distribution

df_cat['City'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Distribution of Cities')
plt.xlabel('City')
plt.ylabel('Count of Transactions')
plt.xticks(rotation=45)
plt.show()
```

# 5) Conclusion  

Key Findings from Descriptive Analytics and Data Visualizations
Statistical Measures:

Mean and Median: Observing the mean and median of numerical columns provided insights into the central tendency of the data. In cases where the mean was significantly different from the median, this indicated potential skewness in the distribution, suggesting the presence of outliers.
Standard Deviation: High standard deviation values in certain columns indicated greater variability among the data points, suggesting diverse customer behaviors or pricing strategies.
Distribution Analysis:

Histograms: The histograms revealed the distributions of numerical variables. For example, a right-skewed distribution may indicate that most sales are concentrated at lower values, with a few high-value transactions affecting the overall mean.
Boxplots: Boxplots highlighted potential outliers in several numerical variables. Extreme values in sales or discounts could be the result of promotional campaigns or errors in data entry.
Categorical Data Insights:

Bar Charts: The frequency distribution of categorical variables showed trends in customer demographics or product categories. For instance, identifying a predominant category in a variable (like high sales in a specific product category) could inform marketing strategies.
Importance of Data Preprocessing Steps
Standardization:

Enhanced Model Performance: Standardizing numerical variables ensures that each feature contributes equally to the distance calculations used in many machine learning algorithms, especially those sensitive to feature scales (e.g., KNN, SVM). This prevents features with larger ranges from dominating the model’s performance.
Improved Interpretability: By transforming data into z-scores, standardized variables facilitate comparison across different features, making it easier to understand relationships and effects.
One-Hot Encoding:

Facilitating Categorical Data Usage: One-hot encoding transforms categorical variables into a numerical format that machine learning algorithms can process. This allows the model to recognize distinct categories as separate features rather than imposing an arbitrary ordinal relationship.
Reducing Bias: One-hot encoding helps to eliminate any potential bias that could arise from label encoding, where the algorithm might misinterpret ordinal relationships between categories that are not inherently ranked.
